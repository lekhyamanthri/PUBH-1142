#### Air pollution, over the course of many decades, specifically in countries like Chad, Bangladesh, India, Nepal and Tajikistan (countries with a higher population), continues to remain prevalent and cause health issues among individuals. Air pollution occurs when there is a presence of one or more contaminants in the atmosphere, such as dust, fumes, gas, mist, odor, smoke or vapor, in quantities that can be injurious to human health. Inhaling these pollutants can lead to inflammation, oxidative stress, immunosuppression and mutagenicity in cells throughout the body, eventually impacting the organs and ultimately leading to disease and hospitalization. The objective of this project is to analyze air pollution levels and their distribution, and use data visualizing techniques to provide valuable insights into the urgent need for cleaner air and healthier cities.


# <p style="color:Green; font-size:30px; font-style: oblique ">***Methods***

##### The data set I used for this project is named: "Air Quality and Health Impact Dataset", and the source is "Kaggle". The dataset contains information on the air quality on public health for 5,811 records. It includes variables such as air quality index (AQI), concentrations of various pollutants, weather conditions and health impact metrics. The target variable is the health impact class, which categorizes the health impact based on the air quality and other related factors. 

# <p style="color:Red; font-size:30px; font-style: oblique ">***Dataset Column Explanation***


##### Record ID: A unique identifier assigned to each record (1 to 2392). 

# <p style="color:Green; font-size:30px; font-style: oblique ">***Air Quality***

##### AQI: Air Quality Index, a measure of how polluted the air currently is or how polluted it is forecast to become. 
##### PM10: Concentration of particulate matter less than 10 micrometers in diameter (μg/m³).
##### PM2_5: Concentration of particulate matter less than 2.5 micrometers in diameter (μg/m³).
##### NO2: Concentration of nitrogen dioxide (ppb). 
##### SO2: Concentration of sulfur dioxide (ppb). 
##### O3: Concentration of ozone (ppb).


# <p style="color:Red; font-size:30px; font-style: oblique ">***Weather Conditions***

##### Temperature: Temperature in degrees Celcius 
##### Humidity: Humidity percentage 
##### Wind Speed: Wind speed in meters per second (m/s)


# <p style="color:Green; font-size:30px; font-style: oblique ">***Health Impact Metrics***

##### Respiratory Cases: Number of respiratory cases reported.
##### Cardiovascular Cases: Number of cardiovascular cases reported.
##### Hospital Admissions: Number of hospital admissions reported.


# <p style="color:Red; font-size:30px; font-style: oblique ">***Health Impact Class (Target Variable)***


##### Health Impact Score: A score indicated the overall health impact based on air quality and other related factors, ranging from 0 to 100. 

##### Health Impact Class: Classification of the health impact based on the health impact score:
##### 0: 'Very High' (Health Impact Score >= 80)
##### 1: 'High' (60 <= Health Impact Score < 80)
##### 2: 'Moderate' (40<= Health Impact Score < 60)
##### 3: 'Low' (20 <= Health Impact Score < 40)
##### 4: 'Very Low' (Health Impact Score < 20)

##### The t-statistic is a measure of the difference between the two groups, while the p-value indicates the probability of observing such a difference by chance. In this case,since the p value is greater than 0.05, we can conclude that the differences are likely due to chance, and there is no significant correlation between hospital admissions and outdoor temperature.
