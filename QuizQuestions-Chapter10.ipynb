```python
import seaborn as sns
```

#### Question 2: 

```python
df = sns.load_dataset("penguins")
df
```

```python
# Create a scatter plot using seaborn
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g")
```

#### Question 3: 

```python
# Create a line plot using seaborn
sns.lineplot(data=df, x="flipper_length_mm", y="body_mass_g")
```

#### Question 4: 


```python
# Create a histogram in seaborn for variable "species"
sns.histplot(data=df, x="species")
```

#### Question 5: 

```python
# Create a stacked histogram in seaborn for variable "species" and 'hue' y in df
sns.histplot(data=df, x="species", hue="species")
```

#### Question 6:

```python
# Create a heatmap in seaborn for a variables island, species in df

sns.heatmap(data=df.pivot_table(index="island", columns="species", values="body_mass_g", aggfunc="mean"))

```

#### Question 7: 

```python
# Create a boxplot in seaborn for variable "species" and 'hue' y in df
sns.boxplot(data=df, x="species", y="body_mass_g")
```

#### Question 8:

```python
# Create a violin plot in seaborn for variable "species" and 'hue' y in df
sns.violinplot(data=df, x="species", y="body_mass_g")
```

#### Question 9: 

```python
# Create a bar plot in seaborn for variable "species" and 'hue' y in df
sns.barplot(data=df, x="species", y="body_mass_g")
```

#### Question 10:

```python
# Add a title to a Seaborn plot assigned to the variable ax with the title 'My Title'
ax = sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g")
ax.set_title("My Title")

```

#### Question 11:

```python
# Save a Seaborn plot ax to a file named 'output.png'
ax.get_figure().savefig("output.png")
```

#### Question 12:

```python
# Set the x and y axis labels to 'X Label' and 'Y Label' respectively for a Seaborn plot ax
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")


```

#### Question 14:

```python
#display a Seaborn plot with a logarithmic scale on the y-axis
import matplotlib.pyplot as plt
ax = sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g")
ax.set_yscale("log")
plt.show()


```

#### Question 15:

```python
# Add a regression line to a scatter plot in Seaborn for variables x and y in a DataFrame df
sns.regplot(data=df, x="flipper_length_mm", y="body_mass_g")
# Add a regression line with confidence interval to a scatter plot in Seaborn for variables x and y in a DataFrame df
sns.regplot(data=df, x="flipper_length_mm", y="body_mass_g", ci=95)

```
